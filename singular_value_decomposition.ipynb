```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

## 1.1 Read the following image (zebra.jpg) into a Python notebook

```python
img2 = cv2.imread('zebra1.jpg')
cv2.imshow('Original',img2,)
plt.imshow(img2)
plt.axis('off')
plt.show()
```

## 1.2 Convert the colored image into a gray level image

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
#cv2.imshow('Grayscale', grayscale)
plt.imshow(img2, cmap = 'gray')
plt.axis('off')
plt.show()
```

## 1.3 Show how many unique elements does image matrix from (1.2) have

```python
len(np.unique(img2))
```

## 1.4 Perform the SVD on the image-matrix of the gray-level image

```python
U, Sigma,V = np.linalg.svd(img2,full_matrices=False)
print(U.shape, Sigma.shape, V.shape)
U@np.diag(Sigma)@V
```

## 1.5 Approximate gray level image matrix by its SVD from (iv) taking different numbers ofsingular values to display the (a) 30% (b) 50% (c) 80% compressions of the image, respectively

$\frac{30}{100*Sigma}= \frac{30}{100*866}= 260$ 

```python
rec = U[:, :260]@np.diag(Sigma[:260])@np.matrix(V[:260, :])
plt.imshow(rec, cmap='gray')
plt.axis('off')
plt.show()
```

$\frac{50}{100*Sigma}=\frac{50}{100*866} = 433$ 

```python
rec = U[:, :433]@np.diag(Sigma[:433])@np.matrix(V[:433, :])
plt.imshow(rec, cmap='gray')
plt.axis('off')
plt.show()
```

$\frac{80}{100*Sigma}= \frac{80}{100*866}= 692$ 

```python
rec = U[:, :692]@np.diag(Sigma[:692])@np.matrix(V[:692, :])
plt.imshow(rec, cmap='gray')
plt.axis('off')
plt.show()
```
